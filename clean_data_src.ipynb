```python
def clean_df_from_path(path):
    # read the file, correcting for spaces after commas, and skipping the index column
    df = pd.read_csv(path, skipinitialspace=True, index_col='Index')
    # lower the column name for queries
    df.columns = df.columns.str.lower()
    # delete Index 
    df.reset_index(drop=True, inplace=True)
    # remove tabulations from the movie column
    df.movie = df.movie.str.replace('\t', '')
    # delete quotes from movies and name columns
    df.name  = df.name.str.replace('"', '')
    df.movie = df.movie.str.replace('"', '')
    # 2. remove one from the year
    df.year -= 1
    
    return df
```

```python
df_male = clean_df_from_path('../data/oscar-age-male-raw.csv')
```

```python
df_female = clean_df_from_path('../data/oscar-age-female-raw.csv')
```

```python
# add data from 2016 for female oscar

latest_female_oscars = pd.DataFrame({
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'age': [28, 60, 44, 50, 63, 44, 60],
    'name': ['Emma Stone', 'Frances McDormand', 'Olivia Colman', 'Renée Zellweger', 'Frances McDormand', 'Jessica Chastain', 'Michelle Yeoh'],
    'movie': ['La La Land', 'Three Billboards Outside Ebbing, Missouri', 'The Favourite', 'Judy', 'Nomadland', 'The Eyes of Tammy Faye', 'Everything Everywhere All at Once']
})
df_female = pd.concat([df_female, latest_female_oscars])
df_female
```

```python
latest_male_oscars = pd.DataFrame({
    'year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'age': [41, 59, 37, 45, 83, 53, 54],
    'name': ['Casey Affleck', 'Gary Oldman', 'Rami Malek', 'Joaquin Phoenix', 'Anthony Hopkins', 'Will Smith', 'Brendan Fraser'],
    'movie': ['Manchester by the Sea', 'Darkest Hour', 'Bohemian Rhapsody', 'Joker', 'The Father', 'King Richard', 'The Whale']
})
df_male = pd.concat([df_male, latest_male_oscars])
df_male
```

```python
# We can't use the syntax df.gender because the field doesn't exist yet
df_male['gender'] = 'M'
df_female['gender'] = 'F'
```

```python
df_concat = pd.concat([df_male, df_female]).sort_values(by='year')
```

```python
df_concat.to_csv('../data/oscar_age_cleaned.csv', index=False)
```
